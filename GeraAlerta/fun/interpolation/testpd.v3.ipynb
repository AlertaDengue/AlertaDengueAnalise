Code to test the functionality of smoothdata.py and missingdata.py

Takes a table of real values and randomly removes entries.
Compare original values with smooth (from full set and from edited one) and interpolated functions.

Python packages needed:
seaborn (can be safely removed, is just used for better graphs)
matplotlib
csv

------------------------
Import libraries (ignore openpyxl warning if it appears. It's a known Panda's v0.14.0 bug):

```python
import copy
import dateutil
try: 
    import seaborn as sb
except:
    pass
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from missingdatapd import fill_missingdata, smoothdata, interpolatenans
from collections import defaultdict
from numpy.random import random_integers
from random import sample
%matplotlib inline
```

Load sample data and extract relevant information:

```python
emptyvals = ['NA','na','Null','null',
             'none','None',None]
fname = 'sampledata/dadosAPS_201517.csv'
sep = ','
dec = '.'
xcol = 'data'
ycol = 'tmin'
df = pd.read_csv(fname, sep=sep, decimal=dec,
                 na_values=emptyvals, keep_default_na=True,
                 usecols=[xcol,ycol,'APS'], parse_dates=[xcol], keep_date_col=True)
df.head()
```

```python
print(df['tmin'].dtype)
print(df['data'].dtype)
```

Obtain smooth function for each AP, from the original data:

```python
df.ix[df['APS']=='AP1',[xcol,ycol]].head()
```

```python
df.ix[:,'t_smooth'] = 0.
df.ix[:,'smooth'] = 0.
apsid = pd.unique(df['APS'])
for k in apsid:
    temp = smoothdata(df.ix[df['APS']==k, [xcol,ycol]], 3)
    df.ix[df['APS']==k, 't_smooth'] = temp.ix[:,xcol].copy()
    df.ix[df['APS']==k, 'smooth'] = temp.ix[:,ycol].copy()
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,xcol], df.ix[df['APS']==k,ycol], label='Original data')
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,'t_smooth'], df.ix[df['APS']==k,'smooth'], label='Smoothed data')
```

```python
plt.clf()
k = 'AP5.1'
plt.plot(df.ix[df['APS']==k,xcol], df.ix[df['APS']==k,ycol], label='Original data')
```

```python
plt.close()
k = 'AP5.1'
plt.plot(df.ix[df['APS']==k,'t_smooth'], df.ix[df['APS']==k,'smooth'], label='Smoothed data')
```

```python
df['tmin_holes'] = df.ix[:,ycol]
df.head()
r = {}
for k in apsid:
    s = df[df['APS']==k].index
    r[k] = sorted(sample(range(min(s), max(s)+1), int(.2*len(s))))
    df.ix[r[k],'tmin_holes'] = np.nan
    temp1, temp2 = fill_missingdata(df.ix[s,[xcol,'tmin_holes']],3)
    df.ix[s,'filled'] = temp1.ix[s,'tmin_holes']
    df.ix[s,'smooth2'] = temp2.ix[s,'tmin_holes']
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'tmin'], label='Original data')
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'smooth'], label='Smooth', lw=3)
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'tmin'], label='Original data')
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'tmin_holes'], label='Sampled data', lw=1)
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'smooth2'], label='Smoothed sampled data', lw=3)
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'tmin_holes'], label='Sampled data', lw=1)
```

```python
plt.close()
k = 'AP1'
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'filled'], label='Estimated data', lw=3)
plt.plot(df.ix[df['APS']==k,'data'], df.ix[df['APS']==k,'tmin'], label='Original data', lw=1)
```

```python
var_orig = {}
var_holes = {}
var_estimate = {}
for k in apsid:
    print(k)
    temp = df.ix[df['APS']==k, 'tmin']
    temp = temp.rename(index={v:v+1 for v in temp.index})
    
    var_orig[k] = temp - df.ix[df['APS']==k, 'tmin']
    medianorig, p90orig, p95orig = np.percentile(abs(var_orig[k]), [50, 90, 95])
    print('Full set')
    print('Median, p90, p95')
    print(medianorig,p90orig,p95orig)

    var_holes[k] = temp.ix[r[k]] - df.ix[r[k], 'tmin']
    medianholes, p90holes, p95holes = np.percentile(abs(var_holes[k]), [50, 90, 95])
    print('Holes')
    print('Median, p90, p95')
    print(medianholes,p90holes,p95holes)

    var_estimate[k] = df.ix[r[k], 'filled'] - df.ix[r[k], 'tmin']
    medianest, p90est, p95est = np.percentile(abs(var_estimate[k]), [50, 90, 95])
    print('Window = ',3)
    print('Median, p90, p95')
    print(medianest,p90est,p95est)
    
    YOut90Orig = sum(abs(var_estimate[k]) > p90orig)
    YOut90Holes = sum(abs(var_estimate[k]) > p90holes)
    YOut95Orig = sum(abs(var_estimate[k]) > p95orig)
    YOut95Holes = sum(abs(var_estimate[k]) > p95holes)
    print('Estimates with error outside p90 and p95, and corresponding percentage of total estimated values')
    print('Full set')
    print(YOut90Orig, str(YOut90Orig*100./len(r[k]))+'%', YOut95Orig, str(YOut95Orig*100./len(r[k]))+'%')
    print('Holes')
    print(YOut90Holes, str(YOut90Holes*100./len(r[k]))+'%', YOut95Holes, str(YOut95Holes*100./len(r[k]))+'%')
    print('')
```
