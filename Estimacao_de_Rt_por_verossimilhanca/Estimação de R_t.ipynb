Acima temos a série de casos, que pretendemos aplicar na função de verossimilhança para $R_K$

A função de verossimilhança é:

$$l = \prod_{k=1}^{K-1}(R_k J_k)^{J_{k+1}} e^{-R_k*J_K}$$

Onde $R_K$ é o numero reprodutivo na semana $K$ e $J_K$ é o número de casos na semana $K$.

logo, a log-verossimilhança é:

$$L = \sum_{k=1}^{K-1} (log(J_k)+log(R_k))J_{k+1}−J_k R_k$$

```python
def lik_R_t(r, K):
    """
    Log-verossimilhança de r[k]
    :param r: lista com rs anteriores
    :param K: número de semanas 
    """
    L = 0
    for k in range(K):
        L += (log(j[k])+log(r[k]))*j[k+1] - j[k]*r[k]
    return L
```

Vamos calcular a verissimilhança de  $R_t$ considerando o suporte $(0,  6)$. Logo, para a primeira semana, $K=1$, podemos usar a função acima para calcular a verissimilhança e o valor que a maximiza.

```python
r_support = arange(0,3,0.01)
r1 = [lik_R_t([r],1) for r in r_support]
```

Seja $\hat{R_K}$ o estimador de maxima verossimilhança de $R_K$:

```python
plot(r_support,exp(r1));
title(u"Verossimilhança de $R_1$");
r1_mle = r_support[r1.index(max(r1))]
plot(r1_mle, exp(max(r1)),'ro');
grid();
xlabel(r'$\hat{R_1} = %s$'%(r1_mle));
```

```python
r2 = [lik_R_t([r1_mle,r],2) for r in r_support]
r2_mle = r_support[r2.index(max(r2))]
plot(r_support,exp(r2));
plot(r2_mle, exp(max(r2)),'ro');
title(u"Verossimilhança de $R_2$");
grid()
xlabel(r'$\hat{R_2} = %s$'%r2_mle);
```

```python
rhats = [r1,r2]
for k in range(3,len(j)):
    past_rs = [r_support[rhat.index(max(rhat))] for rhat in rhats]
    rhats.append([lik_R_t(past_rs+[r],k) for r in r_support])
```

Vamos agora comparar as estimativas de $R_t$ via Maxima Verossimilhança com o método do artigo:

```python
dados['rtmle'] = [0,0]+[r_support[rhat.index(max(rhat))] for rhat in rhats]
dados.rtmle[2:].plot(label="rtmle")
dados.Rt[2:].plot("$R_t$");
legend()
len([r_support[rhat.index(max(rhat))] for rhat in rhats])
```

##Calculating the 95% intervals
For this, we must normalize the likelihood functions, so that they integrate to one, and find the confidence bands using cumsum.

```python
def conf_band(pdf, band=.95):
    #pdf  = pdf[2:]/nansum(pdf[2:])
    pdf = exp(pdf)
    pdf /= nansum(pdf)
    cdf = cumsum(pdf)
    lb = abs(0.025 - cdf)
    ub = abs(0.975 - cdf)
    #print pdf,lb, cdf
    l = r_support[2:][lb.tolist().index(nanmin(lb))]
    u = r_support[2:][ub.tolist().index(nanmin(ub))]
    return l,u
```

```python
bands = array([conf_band(r) for r in rhats])
```
